<p><h1>Standardize the data</h1> :- to increase accuracy

```python
from sklearn.preprocessing import StandardScaler
```

No need to standardize Y_train, Y_test as they are already categorical data (0,1)

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) # Fitting X_train into StandardScaler and transforming it:
X_test_std = scaler.transform(X_test) # Only Transforming X-test data
```

```python
print(X_train_std)
```

Now we Should Use Standardize data for neural network building and training in order to achive more accuracy:

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
# Keras use tensorFlow as backend
```

```python
# Setting up layers of DNN
model = keras.Sequential(
     [
         keras.layers.Flatten(input_shape=(30,)), # convert into 1d array => Input(30 neurons => Features)
         keras.layers.Dense(20,activation='relu'), # all neurons will have activation F(x)
         keras.layers.Dense(30,activation='sigmoid'),
         keras.layers.Dense(2,activation='sigmoid') # number of clases => Output
     ]
 )
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# ReTraining the NN for better accuracy using standardize data
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=15)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('DNN Train Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training data','validation data'] , loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('DNN Loss Accuracy')
plt.legend(['training data','validation data'] , loc = 'lower right')
```

Accuracy of model on test data

```python
loss,accuracy = model.evaluate(X_test_std,Y_test) # here model will evaluate X-test_std data and try to predict the label ,
#    and this label will be compared with Y-test label and it will return accuracy and loss value
```

```python
Y_pred.shape
print(Y_pred[0]) # --> Gives two value prob label being 0 and 1 respectively
```

Y_pred gives the probability of each class so we want to convert those prob into labels(0,1) logic for that is implemented below

```python
# Brute Force
dem_list=[]
for i in Y_pred:
  temp= max(i)
  if(temp == i[0]):
    dem_list.append(0)
  else:
    dem_list.append(1)

print(dem_list)

```

```python
# Using argmax function of numpy

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

```python
dem_list == Y_pred_labels # hence our logic is correct
```

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
